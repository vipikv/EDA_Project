<p style="color: darkblue;font-size:medium">Percenatge of employees in each group</p>

```python
percentage_split = (df['Team'].value_counts()/len(df))*100
percentage = percentage_split.round(2)
percentageDF = pd.DataFrame({'Team':df['Team'].value_counts(),'percentage split':percentage})
percentageDF
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=percentageDF['Team'].index, y='percentage split', data=percentageDF, palette='tab10',hue=percentageDF['Team'].index)
plt.title('Percentage split')
plt.xlabel('Team')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.show()
```

<p style="color: darkgreen;font-size:medium">Data Story</p>

<p style="color: darkgreen;font-size:medium">From the bar plot based on teams and their respective percentage of employees, team New Orleans Pelicans has the most number of employees.It has Total of 19 employees which is 4.15 percentage of total employees in the company</p>

<p style="color: darkblue;font-size:x-large">Find if there is any correlation between age and salary , represent it visually
</p>

```python
df['Age'].corr(df['Salary'])
```

```python
corr_matrix = df[['Age','Salary']].corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix,annot=True,cmap='rocket')
plt.title('correlation of Age and Salary')
plt.show()
```
